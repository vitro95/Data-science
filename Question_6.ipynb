# Question 6: Plot GameStop Stock Graph

```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
gme = yf.Ticker("GME")
```

```python
gme_share_price_data = gme.history(period ="1y")
```

```python
# Plot GameStop Stock Price
def make_graph(x,y, title = "Graph", xlabel = "x axis", ylabel = "y axis", label_legend = "label"):
    plt.plot(x,y , label= label_legend , color ='b')
    plt.title(title, fontsize =14)
    plt.xlabel(xlabel, fontsize =12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

make_graph(gme_share_price_data.index, gme_share_price_data['Close'],"GameStop Stock Prize 1 Year",
           "Date", "Stock Price (USD)", "GameStop Stock Price")
```
