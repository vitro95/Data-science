# Question 5: Plot Tesla Stock Graph

```python
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
tesla = yf.Ticker("TSLA")
```

```python
tesla_share_price_data = tesla.history(period ="1y")
```

```python
# Plot Tesla Stock Price
def make_graph(x,y, title = "Graph", xlabel = "x axis", ylabel = "y axis", label_legend = "label"):
    plt.plot(x,y , label= label_legend , color ='b')
    plt.title(title, fontsize =14)
    plt.xlabel(xlabel, fontsize =12)
    plt.ylabel(ylabel, fontsize=12)
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

make_graph(tesla_share_price_data.index, tesla_share_price_data['Close'],"Tesla Stock Prize 1 Year",
           "Date", "Stock Price (USD)", "Tesla Stock Price")
```
